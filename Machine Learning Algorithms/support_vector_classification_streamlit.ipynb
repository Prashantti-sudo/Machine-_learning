```python
%%writefile prashant_file.py
import streamlit as st
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Title
st.title("Support Vector Classification with Streamlit")

# Upload Dataset
uploaded_file = st.file_uploader("Upload a CSV file", type=["csv"])
if uploaded_file:
    data = pd.read_csv(uploaded_file)
    st.write("Dataset Preview:", data.head())

    # Select Features and Target
    features = st.multiselect("Select Features", data.columns)
    target = st.selectbox("Select Target", data.columns)

    if features and target:
        X = data[features]
        y = data[target]



        # Scaling
        scaler = StandardScaler()
        X = scaler.fit_transform(X)

        # Train-Test Split
        test_size = st.slider("Test Size (as %)", 10, 50, 20) / 100
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

        # Train Model
        model = SVC()
        model.fit(X_train, y_train)

        # Input values
        st.write("Enter values for prediction:")
        inputs = []
        for i in range(len(features)):
            value = st.number_input(f"Enter value for {features[i]}")
            inputs.append(value)

        if st.button("Predict"):
            prediction = model.predict([inputs])
            st.success(f"Prediction: {prediction[0]}")
```

```python
data = sns.load_dataset("titanic")
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
for i in data.columns:
  if data[i].dtype == 'object'or data[i].dtype == "bool" or data[i].dtype == "category":
    data[i] = lb.fit_transform(data[i])
```

```python
data
```

```python
x = data.drop("survived",axis=1)
y = data["survived"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
%%writefile prashant_file.py
import streamlit as st
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.model_selection import train_test_split
import pandas as pd

# Title
st.title("Naive_Bayes")

# Upload Dataset
uploaded_file = st.file_uploader("Upload a CSV file", type=["csv"])
if uploaded_file:
    data = pd.read_csv(uploaded_file)
    st.write("Dataset Preview:", data.head())

    # Select Features and Target
    features = st.multiselect("Select Features", data.columns)
    target = st.selectbox("Select Target", data.columns)

    if features and target:
        X = data[features]
        y = data[target]



        # Scaling


        # Train-Test Split
        test_size = st.slider("Test Size (as %)", 10, 50, 20) / 100
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

        # Train Model
        model3 = BernoulliNB()
        model3.fit(X_train, y_train)



        # Input values
        st.write("Enter values for prediction:")
        inputs = []
        for i in range(len(features)):
            value = st.number_input(f"Enter value for {features[i]}")
            inputs.append(value)

        if st.button("Predict"):
            prediction = model3.predict([inputs])
            st.success(f"Prediction: {prediction[0]}")
```
